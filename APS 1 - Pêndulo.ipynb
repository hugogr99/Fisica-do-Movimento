```python
#Bibliotecas necess√°rias
import matplotlib.pyplot as plt
import numpy as np
from math import *
from scipy.integrate import odeint
```

```python
#lista de tempo
t = np.arange(0,8,0.01)

#Par√¢metros
g = 9.8 #m/s^2
l = 1.5 #m
m = 0.05 #kg

#Equa√ß√µes diferenciais
def EqDif(solucao, t):
    Œ∏ = solucao[0]
    œâ = solucao[1]
    dŒ∏dt = œâ
    dœâdt = -(g*sin(ùúÉ))/l
    return [dŒ∏dt,dœâdt]

#Condi√ß√µes iniciais
Œ∏_inicial = 20
œâ_inicial = 0
ci = [radians(Œ∏_inicial),œâ_inicial]
```

```python
#Rodando o ODEINT
solucao = odeint(EqDif,ci,t)
```

```python
#Para facilitar a modelagem, os eixos x e y foram invertidos, ou seja, x refere-se ao eixo das coordenadas e y ao das abcissas

x = []
y = []
Tr = []
ang = []

for i in range(0,len(solucao[:,0])):
    ùúÉ=(solucao[:,0])[i]
    ang.append(ùúÉ)
    ùúî=(solucao[:,1])[i]
    T= m*(ùúî**2)*l+m*g*cos(ùúÉ)
    Tr.append(T)
    x.append(l*cos(ùúÉ))
    y.append(l*sin(ùúÉ))
```

```python
#Gr√°fico xy
plt.plot(y,x)
plt.axis([min(y),max(y), max(x), min(x)])
plt.axis('equal')
plt.title('Posi√ß√£o do peso nos eixos x e y no tempo')
plt.xlabel('x (m)')
plt.ylabel('y (m)')
plt.grid(True)
plt.show()

#Gr√°fico posi√ß√£o em y no tempo
plt.plot(t,x)
plt.title('Posi√ß√£o de y no tempo')
plt.xlabel('Tempo (s)')
plt.ylabel('y (m)')
plt.grid()
plt.show()

#Gr√°fico de posi√ß√£o x no tempo
plt.plot(t,y)
plt.title('Posi√ß√£o de x no tempo')
plt.xlabel('Tempo (s)')
plt.ylabel('x (m)')
plt.grid()
plt.show()

#Gr√°fico de varia√ß√£o angular no tempo
plt.plot(t,ang)
plt.title('Varia√ß√£o do √¢ngulo no tempo')
plt.xlabel('Tempo (s)')
plt.ylabel('√Çngulo medido entre o eixo y e o peso (rad)')
plt.grid()
plt.show()
```
